```python
count_dict = {}
for i in train_array:
    num_slot = len(re.findall(':B',str(i)))
    if num_slot in count_dict.keys():
        count_dict[num_slot] += 1
    else:
        count_dict[num_slot] = 1
        print("Slots:",num_slot,i)
     
```

```python
intent_count = {}
for i in train_array:
    intent = re.findall('(?<=<=> )\w+',str(i))
    if intent[0] in intent_count.keys():
        intent_count[intent[0]] += 1
    else:
        intent_count[intent[0]] = 1
```
