```python
import pandas as pd
import numpy as np 
import re
import random
import copy
```

```python
file_name = 'appliance_data\\appliance_utterances.xlsx'
data = pd.read_excel(file_name)
```

```python
data.head()
```

```python
data.shape[0]
```

```python
train = {"edge": [], "question_should_contain": [], "question": [], "intent":[]}
test = {"edge": [], "question_should_contain": [], "question": [], "intent":[]}
valid = {"edge": [], "question_should_contain": [], "question": [], "intent":[]}
```

```python
for i in range(data.shape[0]):
    if ((i % 4 == 0) or (i % 4 == 1)):
        train['edge'].append(data.loc[i,'edge'])
        train['question_should_contain'].append(data.loc[i,'question_should_contain'])
        train['question'].append(data.loc[i,'question'])
        train['intent'].append(data.loc[i,'intent'])
    elif (i % 4 == 2):
        test['edge'].append(data.loc[i,'edge'])
        test['question_should_contain'].append(data.loc[i,'question_should_contain'])
        test['question'].append(data.loc[i,'question'])
        test['intent'].append(data.loc[i,'intent'])
    else:
        valid['edge'].append(data.loc[i,'edge'])
        valid['question_should_contain'].append(data.loc[i,'question_should_contain'])
        valid['question'].append(data.loc[i,'question'])
        valid['intent'].append(data.loc[i,'intent'])
```

```python
df_train = pd.DataFrame.from_dict(train)
df_test = pd.DataFrame.from_dict(test)
df_valid = pd.DataFrame.from_dict(valid)
```
