```python
file = open('C:\\Users\\l.kwon\\slot_filling_and_intent_detection_of_SLU\\exp\\model2\\valid.iter5')
```

```python
def extract_question(array):
    sentence = []
    for i in array:
        a = i.split(":")[0]
        sentence.append(a)
    return ' '.join(sentence)
```

```python
slot = {}
intent = {}
```

```python
k = 0
for x in file:
    wrong = False
    array = x.split(" <=> ")
    if(' '.join(array[1].split()) != ' '.join(array[2].split())):
        print(k,"Intent Match Failed:",array[1],array[2])
        if array[1] in intent.keys():
            intent[array[1]] += 1
        else:    
            intent[array[1]] = 1
        if array[2] in intent.keys():
            intent[array[2]] += 1
        else:    
            intent[array[2]] = 1
        wrong = True
    array_1 = array[0].split(" ")
    question = extract_question(array_1)
    for i in array_1:
        array_2 = i.split(':')
        if(array_2[1] != array_2[2]):
            print(k,array_2[0],array_2[1],array_2[2])
            wrong = True
    k+=1
    if(wrong == True):
        print(question)
        print("------------------------------------------------")
```
